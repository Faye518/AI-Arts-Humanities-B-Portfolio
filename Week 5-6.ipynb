# Week 5–6 Development (MS Euing 26)

**Object:** MS Euing 26 (14th-century Carmelite missal, decorated initials)  
**Goal:** Explore how generative AI can reinterpret illuminated initials while reflecting on authenticity and human–AI collaboration.

## Inputs (digitised material from ASC)

The following images were photographed during my visit to the Archives and Special Collections (ASC).
They capture different aspects of MS Euing 26, including decorated initials, page layout, colour palette, and material qualities such as parchment and gilded edges.

These images form the primary input for my generative experiments.

## Generation experiment – First attempt

**Intention:**  
To generate a medieval-style illuminated initial inspired by MS Euing 26, focusing on colour (gold, blue, red) and decorative borders.
