# Week 2–4 Log (Exploration → Direction)

## Week 2 – Tools & first exploration
- Tools I tried: Chatgpt，Microsoft Copilot
- What I asked / tested:I asked the AI to summarise the historical background of MS Euing 26 and MS Gen 13.
- What I got:The AI provided a general overview of medieval manuscripts and travel journals.
- Quick fact-check:After reviewing the collection website of the University of Glasgow, I found that the AI's account was a bit too general in some details and did not match the actual archival items.

## Week 3 – Small experiment
- What I generated (text/image/audio):During my visit to the Archives and Special Collections (ASC) facility, I carefully observed these physical items and made detailed records of their materials, textures, and layouts.
- What I changed (prompt/choices):I no longer relied solely on the summaries generated by the AI, but compared the AI's description with my own direct observations.
- What I learned about human–AI collaboration:I realized that AI cannot capture the material and sensory aspects of archival objects. Interpreting their specific physical details and background situations requires human observation.

## Week 4 – Project direction (early idea)
- Possible object/topic:MS Euing 26 – a 14th century Carmelite missal with decorated initials.
- Why it matters (value/context):The manuscript reflects medieval religious culture and visual aesthetics.
- What I want the audience to notice/learn:I want to explore how generative AI can recreate or reinterpret illuminated initials and question the notion of authenticity.
- Where AI will help (and where it won’t):AI cannot replace the historical context or material authenticity of the original manuscript.

## Reflection (short)
- What AI is good at:Generating stylistic variations quickly.
- What AI struggles with:Understanding historical nuance and material authenticity.
- What I (human) contributed:Critical judgement, selection, contextual interpretation.
